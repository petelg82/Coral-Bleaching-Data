# **Part 4 - Filtering the Atlantic Data**
### _Purpose: This code will take the raw global bleaching data and separate all the data from the Atlantic_
## **Step 1** - Type your given nickname for the global bleaching csv file from above and type it followed by an outward facing square bracket
## **Step 2** - Type your given nickname for the global bleaching csv file from above again followed by another outward facing square bracket
## **Step 3** - In double Quotation marks type _Ocean_Name_ which will isolate the data from the Ocean_Name column of the raw dataset

# **Part 5 - Putting the Atlantic Ocean data into a subset**
### _Purpose: This part will allow you to manipulate just the data from the Atlantic Ocean._
## **Step 1** - Create a nickname for the subset of the data and then type an underscore followed by the word _Subset_ (I chose to name it Atlantic_Subset
### _**Note**: You cannot include any spaces, you have to type an underscore instead_
## **Step 2** - Type a space followed by a single equal sign and another space
## **Step 3** - Copy the exact code from above after the space

```python
Atlantic_Subset = turtle[turtle["Ocean_Name"]=="Atlantic"]
```

# **Part 6 - Create another subset that filters the data based on the mean temperature from each site**
### _Purpose: This part will give you another subset of data that contains only the site that have a mean temperature over 302.15 degrees Kevlin or 29 degrees Celsius, which is the maximum temperature most coral can survive in._
## **Step 1** - Create a nickname for this second subset followed by an underscore and the word _Subset_ (I called mine TooHotAtlantic_Subset)
## **Step 2** - Type a space then an equal sign followed by another space
## **Step 3** - Type the nickname you created for the Atlantic Subset after the last space followed by an outward facing square bracket
## **Step 4** - Type the nickname you created for the Atlantic subset again followed by another outward facing bracket
## **Step 5** - Type a double quotation mark followed by _Temperature_Mean_ which is the name for the column of the Atlantic Subset containing the mean temperature of the site followed by another double quoation mark
## **Step 6** - Type an inward facing bracket followed by greater than sign and an equal sign right after (it should look like this: **>=**)
## **Step 7** - Type a double quotation mark followed by _302.15_ and another double quotation mark
## **Step 8** - Close the code sequence with an inward facing bracket

```python
TooHotAtlantic_Subset = Atlantic_Subset[Atlantic_Subset["Temperature_Mean"]>="302.15"]
```
